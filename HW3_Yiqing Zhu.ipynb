```python
from pandas import Series, DataFrame
import pandas as pd
%pylab inline
```

# Q1 Read the data

Read the two datasets. Drop duplicates for both of them.

```python
# Read the data gold
gold=pd.read_table('gold.txt',names=['URL','Category']).drop_duplicates()
print gold
```

```python
#Read the data label
label=pd.read_table('labels.txt',names=['turk','URL','Category']).drop_duplicates()
print label
```

# Q2 Split into two DataFrames

Split non-gold URLs and gold URLs by defing two functions. Create masks by mapping functions on URL column in label dataset. Apply masks to get two seperate dataframes.

```python
# Get all URLs that appeared in gold dataset
gold_url=gold['URL'].values 

#Create two functions to split non-gold URLs and gold URLs
def in_gold(URL): 
    return URL in gold_url

def not_in_gold(URL):
    return URL not in gold_url
# map label dataset to splic the original data into two
mask1=label['URL'].map(in_gold)
labels_on_gold=label[mask1]
labels_on_gold

mask2=label['URL'].map(not_in_gold)
labels_unknown=label[mask2]
labels_unknown
```

# Q3 Compute accuracies of turks

Merge gold set and labels_on_gold set on the same URL. Add a column named accuracy. Accuracy was given 1 if the rate of a certain turk is correct, otherwise, accuracy was given 0.
Then use groupby fuction to compute counts of ratings and mean of accuracies of each turk.

```python
#merge the two datasets on URL column
gold_turf_merge=pd.merge(labels_on_gold,gold,left_on=['URL'],right_on=['URL'])
gold_turf=gold_turf_merge[['turk','Category_x','Category_y']]
#add a column named accuracy. Accuracy was given 1 if category given by a certain turk for an URL is correct;
#Otherwise, accuracy was given 0.
gold_turf['accuracy']=gold_turf.apply(lambda row: row['Category_x'] in row['Category_y'],axis=1)

#maker accuracy integer
gold_turf['accuracy']= gold_turf['accuracy'].astype(int)
#Use groupby to get the counts of ratings of each turk and the mean of their rating accuracies
rating_counts=gold_turf.groupby('turk')[['Category_x']].count()
rating_average=gold_turf.groupby('turk')[['accuracy']].mean()
#append the rating_average acolumn to the rating_counts dataset to form rater_goodness
rater_goodness_T=rating_counts.T.append(rating_average['accuracy'])

rater_goodness=rater_goodness_T.T
rater_goodness=rater_goodness.rename(columns={'Category_x':'counts'})
rater_goodness


```

# Q4 Odd Ratios

```python
#add a column called odds
rater_goodness['odds']=rater_goodness['accuracy']/(1.001-rater_goodness['accuracy'])
rater_goodness

```

# Q5 Most Accurate turks

Filter out counts of ratings that is above 20. Then sort the dataframe according to accuracy in a descending order. Select the top 10 and return turks ID.

```python
#Create a mask to filter out counts that are above 20.
mask_above20=rater_goodness['counts']>=20
rater_goodness_20=rater_goodness[mask_above20]
#Sort the dataset in an descending order.
rater_goodness_20=rater_goodness_20.sort(columns=['accuracy'],ascending=False)
#get top 10 turks
#rater_goodness_20[:10]
rater_goodness_20.index.tolist()[:10]
```

# Q7 Overall predicted odds

Add a 'turk' column to rater_goodness in order to do the merge. Merge rater_goodness and labels_unkown datasets. Select the top25% turks that rates more gold URLs by using quantile techniques. 
Then use groupby to get the produt of odds for each category of an URL.

```python
#convert the turk index to column
rater_goodness['turk']=rater_goodness.index
#Merge labels_unknown dataset with rater_goodness dataset to filter out turks with odds in the labels_unkown dataset.
turk_in_unknown=pd.merge(labels_unknown,rater_goodness,left_on=['turk'],right_on=['turk'])
#Computer the 75%quantile of counts and use it as a mask to select the top 25% turks.
number_quantile=rater_goodness['counts'].quantile(0.75)
within_25=turk_in_unknown[turk_in_unknown['counts']>number_quantile]
#Take groupby function to get the production of odds for each category of URL
grps=within_25.groupby(['URL','Category'])[['odds']].prod()
grps
```

# Q8 Predicted Categories

Using unstack and transpose in order to apply max and idxmax function to get top category and top odds.

```python
#unstack grps
grps_unstacked=grps.unstack('Category')['odds']
#Transpose unstaked grps and get the max odds and responsible category for each URL
top_odds=grps_unstacked.T.max()
top_category=grps_unstacked.T.idxmax()

#combine the two Series
result_75=pd.concat([top_category,top_odds],axis=1,keys=['top category','top odds'])
result_75
```

# Q9 Predicted Categories using more turks

Repeat Q7 but change 75 quantile to 25. Then replicat Q8.
Use crosstab to get the table matrix.
We can find that the most errors lie in pair(G,P).

```python
#convert the turk index to column. Repeat Q7
rater_goodness['turk']=rater_goodness.index

turk_in_unknown=pd.merge(labels_unknown,rater_goodness,left_on=['turk'],right_on=['turk'])

number_quantile_25=rater_goodness['counts'].quantile(0.25)
within_75=turk_in_unknown[turk_in_unknown['counts']>number_quantile_25]
grps_new=within_75.groupby(['URL','Category'])[['odds']].prod()
grps_new
```

```python
#Repeat Q8
grps_unstacked_new=grps_new.unstack('Category')['odds']

top_odds_new=grps_unstacked_new.T.max()
top_category_new=grps_unstacked_new.T.idxmax()



result_25=pd.concat([top_category_new,top_odds_new],axis=1,keys=['top category','top odds'])
result_25
```

```python
#Create an URL column in each of the dataset
result_25['URL']=result_25.index
result_75['URL']=result_75.index
#Merge the two datasets according to same URLs
merge_75_25=pd.merge(result_75,result_25,on=['URL'],suffixes=('_75','_25'))

#Use crosstab to get the table matrix.
merge_counts=pd.crosstab((merge_75_25['top category_75'] ),merge_75_25['top category_25'])
merge_counts
```
