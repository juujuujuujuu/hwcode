# 2. Voters in Florida

```python
import re
fp=open("FloridaVoters.html","r").readlines()
result=[]
l = []

for line in fp:
#Read through html code and extract table information denoted by <td>. 
#Append all information into one list after deleting space and commas.
    if re.match("<td>", line):
        line=line.strip().rstrip('</td>').lstrip('<td>')
        line=line.replace(",","")
        result.append(line)
n=6

for i in range (0,len(result),n):
#Divides result list into groups of six items (county name and its five corresponding voter totals)
#Add only first 3 items into new list, change items 2,3 to integers
#Change new list to tuple and append to main 'l' list
    newlist=[result[i:i+n]] 
    for county in newlist:
        newlist = county[0:3]
        newlist[1] = int(newlist[1])
        newlist[2] = int(newlist[2])
        newlist = tuple(newlist)
        l.append(newlist)   

#Sorts list by third item (democratic voters)
l.sort(key=lambda tup: tup[2])


for i in range(0,len(l)):
    #prints items in correct order: County, Republic Voters, Democrat Voters
    print l[i][0],l[i][1],l[i][2]
```

# 3. The Google of Quotes

## (a) A list of full quotes

```python
f = open("quotes.txt","r")
l_quotes = []

while True:
    #seperately store one quote to str1 and its speaker to str2
    str1 = f.readline()
    str1 = str1.rstrip()
    str2 = f.readline()
    str2 = str2.rstrip()     
    
    #The While-loop continues as long as str2 is not null
    if str2 == "":
        break       
    else: 
        #combine str1 and str2 with "-", then format and append the result to a full quote list
        str1 = str1 + " - " + str2
        l_quotes.append(str1)            
print l_quotes
```

## (b) Words from full quotes

```python
import re

f = open("quotes.txt","r")
l_words = []

for line in f.readlines():
    #split every line with one or more "\W" (means anything except characters, numbers and underscores)
    words = re.split("\W+",line)
    #remove spaces in the list and lower the cases of words
    for word in words:
        l_words.append(word.lower())        
    while "" in l_words:
        l_words.remove("")
        
print l_words
```

## (c) Build the postings-list dictionary

```python
import re
from collections import Counter

def post(str1):
    """create a dictionary whose keys are full quotes, and whose values are themselves dictionaries with key being a word, and value being the number of times
    the word occurs in the full quote"""
    
    dict1 = {}
    words = re.split("[^a-zA-Z0-9_]+",str1.lower()) # lower the case of the quote and split the quote into word pieces
    Counter(words) # counter function creates a dictionary whose keys are words and whose values are the frequency of the word
    dict1[str1] = dict(Counter(words)) # Add keys(quote) and values(another dictionary) to the dict1 dictionary
    return dict1

print post("I may be drunk, Miss, but in the morning I will be sober and you will still be ugly. - Winston Churchill")
```

## (d) Build the reverse postings-list dictionary

```python
from collections import Counter
import re

def reverse_post(word):
    """ the function creates a dictionary whose keys are the words, and the values are themselves dictionaries with the key 
    being a full quote, and the value being the number of times the word appeared in the full quote."""
    
    # create two empty dictionaries and an empty list
    dict1 = {}
    dict2 = {}
    l_words = []
  
    # l_quotes is a list of all the quotes(defined in question a ), then iterate every single quote in the l_quotes
    for quote in l_quotes:
        if word in quote:         # if the word is in the quote, then append that word to the l_words list
            l_words.append(quote) # Now the l_words contains all the quotes that contain the word
    
    # iterate through every single quote (contains the word) in the list l_words
    for str3 in l_words:
        split_words = re.split("\W+",str3)   # split every single quote into several strings    
        result = dict(Counter(split_words))  # counts the frequency of each word in a string and restore the result in a dictionary
        for key in result:                   # iterate through each key in the result dictionary
            if key == word:            
                dict1[str3] = result[key]    # in the small dictionaryeach, each quote is the key,  and value is the frequency of words appear in the qutoe
                dict2[word] = dict1
    return dict2                   

print reverse_post("Rita")    
```

## (e) Write a TF-IDF function

```python
import math
from collections import Counter

def TF_IDF(word, quote):      
    d={}
    #d gets the word frequency of all words in our input quote
    d = post(quote).values()[0]
    word = word.lower()
    #b gets the frequency of our input word
    b = d[word]
    #a gets the highest frequency of any word occurred in the quote containing our input word
    a = float(max(d.values()))
    #calculate TF
    TF = b/a
    #calculate IDF with the total number of quotes divided by the number of quotes containing our input word
    IDF = math.log(len(l_quotes)/len(reverse_post(word)))
    #calculate TD-IDf
    TF_IDF = TF * IDF 
    return TF_IDF

print TF_IDF("entertainer", 'An actor is at most a poet and at least an entertainer.- Marlon Brando')
```

## (f) Quote search using a single word

```python
def single_word_search(word):
    """create a function that takes a word as argument, and returns a dictionary whose keys are full quotes containing 
    that word, and whose values are the TF-IDF score of that word for that full quote."""
        
    search = {}                              # create an empty dictionary 
    word = word.lower()  
    quotes = reverse_post(word)[word].keys() # call the reverse_post function and obtain all keys of the inside dictionary
    for quote in quotes:                     # iterate through every quote in the quotes list
        search[quote] = TF_IDF(word,quote)   # calculate the TF_IDF value given the word and the quote
    return search
 
print single_word_search("entertainer")
```

## (g) Quote search using multiple words

```python
import re

def multi_word_search(list_words):
    """create a function that takes a list of words as argument, and returns a dictionary whose keys are full quotes 
    containing that word, and whose values are the sum of TF-IDF scores of all the words for that full quote."""
    
    search = {}
    for word in list_words:                # iterate through each word in the list of words
        for key, value in single_word_search(word).items(): # call the single_word_search function and iterate trhough the key, value
            if key not in search.keys():   # if the key(a quote)does not appear in the empty list, add key and value to that dictionary
                search[key] = value        # the key should be a quote, and the value should be the TF_DIF value
            else:
                search[key] += value       # if >=2 words in list_words appear in the same quote then the value should be added up 
    return search                          

print multi_word_search(["entertainer","actor"])
```
